#### Gender
<br>We will start with the columns that have less rows in the dataframe (missing values)

```python
data['Gender'].unique()
```

```python
data['Gender'] = data['Gender'].astype('str')
```

```python
#Gender, standardize values
data = data.replace({'Gender' : {'1':'male', '0':'unknown', '2':'female', 1:'male', 2:'female', 0:'unknown'}})
```

```python
data['Gender'].unique()
```

#### Birth Year

```python
data['Birth Year'].unique()
```

```python
data['Birth Year'] = data['Birth Year'].astype('str')
```

```python
#Turn Birth Year into an integer type
data['Birth Year'] = data['Birth Year'].astype('int32')
```

#### Age (new)

#### Start Time / Start Date

```python
#Check if there are values on Start Time that do not contain a date as we are going to split that column
noStartDate = data[~data['Start Time'].str.contains('/')]
noStartDate
```

```python
# Create a new data frame with split value columns 
new = data['Start Time'].str.split(" ", n = 1, expand = True)
```

```python
new.head()
```

```python
# Change the values of Start Time column to just time
data['Start Time']= new[1] 

# Add a new column for Start Date
data.insert(3, 'Start Date', new[0])
```

```python
data.head()
```

#### Stop Time / Stop Date

```python
#Check if there are values on Stop Time that do not contain a date as we are going to split that column
noStopDate = data[~data['Stop Time'].str.contains('/')]
noStopDate
```

```python
# Create a new data frame with split value columns 
new2 = data['Stop Time'].str.split(" ", n = 1, expand = True)
```

```python
new2.head()
```

```python
# Change the values of Start Time column to just time
data['Stop Time']= new2[1] 

# Add a new column for Start Date
data.insert(5, 'Stop Date', new2[0])
```
